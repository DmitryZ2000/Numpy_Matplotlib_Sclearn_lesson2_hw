##**Задание 1**
Создайте массив Numpy под названием a размером 5x2, то есть состоящий из 5 строк
и 2 столбцов. Первый столбец должен содержать числа 1, 2, 3, 3, 1, а второй - числа 6,
8, 11, 10, 7. Будем считать, что каждый столбец - это признак, а строка - наблюдение.
Затем найдите среднее значение по каждому признаку, используя метод mean

```python
a = np.array([[1, 2, 3, 3, 1],[6, 8, 11, 10, 7]]).T
a
```

```python
mean_a = a.mean(axis=0)
mean_a
```

##**Задание 2**
Вычислите массив a_centered, отняв от значений массива “а” средние значения
соответствующих признаков, содержащиеся в массиве mean_a. Вычисление должно
производиться в одно действие. Получившийся массив должен иметь размер 5x2

```python
a_centered = a - mean_a
a_centered
```

##**Задание 3**
Найдите скалярное произведение столбцов массива a_centered. В результате должна
получиться величина _centered_spa. Затем поделите a_centered_sp на N-1, где N - число наблюдений.

```python
_centered_spa = a_centered[:, 0] @ a_centered[:, 1]
_centered_spa
```

```python
_centered_spa = np.dot(a_centered[:, 0], a_centered[:, 1])
_centered_spa
```

##**Тема “Работа с данными в Pandas”**

Импортируйте библиотеку Pandas и дайте ей псевдоним pd. Создайте датафрейм authors со
столбцами author_id и author_name, в которых соответственно содержатся данные: [1, 2, 3] и
['Тургенев', 'Чехов', 'Островский'].
Затем создайте датафрейм book cо столбцами author_id, book_title и price, в которых соответственно
содержатся данные:
[1, 1, 1, 2, 2, 3, 3],
['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и
поклонники']
Нужны prices!
[500, 400, 300, 350, 450, 600, 200]

```python
import pandas as pd
```

```python
a1 = {'author_id': [1, 2, 3], 'author_name' : ['Тургенев', 'Чехов', 'Островский']}
authors = pd.DataFrame(a1)
authors
```

```python
a2 = {'author_id': [1, 1, 1, 2, 2, 3, 3],
      'book_title':['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
      'price': [500, 400, 300, 350, 450, 600, 200]}
books = pd.DataFrame(a2)
books
```

**Задание 2**  
Получите датафрейм authors_price, соединив дата фреймы authors и books по полю author_id.

```python
authors_price = pd.merge(authors, books, on='author_id', how='inner' ).copy()
authors_price
```

**Задание 3**  
Создайте датафрейм top5, в котором содержатся строки из authors_price с пятью самыми дорогими книгами.

```python
top5 = authors_price.nlargest(5, 'price').copy()
top5
```

##**Задание 4**
Создайте датафрейм authors_stat на основе информации из authors_price. В датафрейме authors_stat должны быть четыре столбца:  
author_name, min_price, max_price и mean_price,  
в которых должны содержаться соответственно имя автора, минимальная, максимальная и средняя цена на книги этого автора.


```python
authors_stat = authors_price.groupby('author_name')
```

```python
authors_stat_max = authors_stat.agg({'price': 'max'}).rename(columns={'price':'max_price'})
authors_stat_max
```

```python
authors_stat_min = authors_stat.agg({'price': 'min'}).rename(columns={'price':'min_price'})
authors_stat_min
```

```python
authors_stat_mean = authors_stat.agg({'price': 'mean'}).rename(columns={'price':'mean_price'})
authors_stat_mean
```

```python
authors_stat = pd.merge(authors_stat_max, authors_stat_min, on='author_name')
authors_stat = pd.merge(authors_stat, authors_stat_mean, on='author_name').reset_index().copy()
authors_stat
```

```python
# Альтернативный способ
authors_stat_2 = authors_price.groupby('author_name')
authors_stat_new = authors_stat_2.agg({'price': 'max'}).rename(columns={'price':'max_price'})
authors_stat_new
```

```python
authors_stat_new["min_price"] = authors_stat_2["price"].min()
authors_stat_new['mean_price'] = authors_stat_2["price"].mean()
authors_stat_new.reset_index()
```

```python
# Еще один вариант
authors_stat = pd.pivot_table(authors_price, values='price', index=['author_name'], aggfunc={'price': [max, min, np.mean]}).reset_index().copy()
authors_stat.rename(columns={'max':'max_price', 'min': 'min_price', 'mean':'mean_price'}, inplace=True)
authors_stat
```
